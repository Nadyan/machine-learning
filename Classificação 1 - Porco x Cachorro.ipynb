```python
from sklearn.svm import LinearSVC

# feaures
# pelo longo ou curto
# perna curta ou longa
# faz au au
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]
```

```python
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1, 1, 1, 0, 0, 0] # 1 - porco, 0 - cachorro (labels)
```

```python
# estimador
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
```

```python
animal_misterioso1 = [1, 1, 1]
animal_misterioso2 = [1, 1, 0]
animal_misterioso3 = [0, 1, 1]

teste_x = [animal_misterioso1, animal_misterioso2, animal_misterioso3]
teste_y = [0, 1, 1] # valores reais (na verdade o terceiro animal Ã© um porco)

previsoes_teste = modelo.predict(teste_x) # previsoes feitas pelo LinearSVC
print(previsoes_teste)
```

```python
corretos = (previsoes_teste == teste_y).sum()
total = len(teste_x)
taxa_acerto = corretos / total
print(corretos, 'de', total, 'corretos')
print("Taxa de acerto: %.2f" % (taxa_acerto*100), '%')
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto = accuracy_score(teste_y, previsoes_teste)
print("Taxa de acerto: %.2f" % (taxa_acerto*100), '%')
```

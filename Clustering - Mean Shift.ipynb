O algoritmo Mean Shift não requer parametrização como o DBSCAN ou o Kmeans.
Cada ponto inicial é considerado um cluster, fazendo os demais pontos convergirem ao ponto de maior densidade.

Vantagens:
 - Funciona com ruído e detecta diversas formas, não apenas radial;

Desvantagens:
 - Não funciona com dados de alta dimensionalidade;
 - Computacionalmente caro.

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/alura-cursos/Clustering/Aula-1.-An%C3%A1lise-Explorat%C3%B3ria/Wine.csv"

df = pd.read_csv(url)
df.head()
```

```python
### Normalização dos dados:

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_normalizado = scaler.fit_transform(df) ## Retorna um numpy Array
df_normalizado = pd.DataFrame(data=df_normalizado, columns=df.keys()) ## Transforma em DataFrame

df_normalizado.describe()
```

```python
from sklearn.cluster import MeanShift

agrupador = MeanShift()
agrupador.fit(df)
agrupador.labels_
```

```python
## Gráfico cluster relacionando a intensidade da cor com a quantidade de alcool de cada grupo:

fig = go.Figure()
fig.add_trace(go.Scatter(x = df['Color_Intensity'], y = df['Alcohol'], 
                            mode = 'markers',
                            marker = dict(color = agrupador.labels_.astype(np.float)),
                            text = agrupador.labels_,
                         ))
fig.update_layout(scene = dict(
    xaxis_title = "Intensidade de cor",
    yaxis_title = "Álcool"
))
fig.show()
```

```python
## Gráfico em 3D:

fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df['Color_Intensity'], y = df['Alcohol'], z = df['Proline'],
                            mode = 'markers',
                            marker = dict(color = agrupador.labels_.astype(np.float)),
                            text = agrupador.labels_
                         ))

fig.update_layout(scene = dict(
    xaxis_title = "Intensidade de cor",
    yaxis_title = "Álcool",
    zaxis_title = "Proline"
))
fig.show()
```

```python
### Podemos variar um parâmetro do algoritmo Mean Shift, o Bandwidth:

from sklearn.cluster import estimate_bandwidth

bw = estimate_bandwidth(df, quantile = 0.1)
bw ## 63.18
```

```python
agrupador = MeanShift(bandwidth=bw)
agrupador.fit(df)
agrupador.labels_
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = df['Color_Intensity'], y = df['Alcohol'], 
                            mode = 'markers',
                            marker = dict(color = agrupador.labels_.astype(np.float)),
                            text = agrupador.labels_,
                         ))
fig.update_layout(scene = dict(
    xaxis_title = "Intensidade de cor",
    yaxis_title = "Álcool"
))
fig.show()
```

Quanto menor o bandwidth, mais clusteres o algoritmo irá encontrar.

 - O quantil define a bandwidth.
     - Quantil é a parte da lista ordenada escolhida, quando menor o quantil, mais próximo do inicio da fila.

Para avaliação do cluster, utilizaremos o Coeficiente de Silhueta:

```python
from sklearn.metrics import silhouette_score

faixa_quantil = np.linspace(0.02, 1, 30) ## vetor com quantis variando de 0.02 até 1
faixa_quantil
```

```python
valores_silhueta = []
quantil_plot = []

for quantil in faixa_quantil:
    bandwidth = estimate_bandwidth(df, quantile = quantil) ## estima a bw para cada quantil
    agrupador = MeanShift(bandwidth)
    labels = agrupador.fit_predict(df)

    if(len(np.unique(labels)) > 1 and len(np.unique(labels)) < len(df)): ## a qtd de labels deve ser maior que 1 e menor que a quantidade de dados
        media_silhueta = silhouette_score(df, labels)
        valores_silhueta.append(media_silhueta)
        quantil_plot.append(quantil) ## lista de quantis válidos
```

```python
quantil_plot ## quantis utilizados na avaliação
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = quantil_plot, y = valores_silhueta))
fig.update_layout(dict(xaxis_title = 'Quantil', yaxis_title = 'Média Silhueta'))
fig.show()
```

A melhor média de silhueta é quando o quantil atinge o valor de 0.35.

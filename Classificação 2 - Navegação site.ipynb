```python
import pandas as pd

url = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
dados = pd.read_csv(url)
dados.head()
```

```python
mapa = {
    "home" : "principal",
    "how_it_works" : "como_funciona",
    "contact" : "contato",
    "bought" : "comprou"
}

dados = dados.rename(columns = mapa)
```

```python
x = dados[["principal", "como_funciona", "contato"]]
y = dados["comprou"]
print(x)
```

```python
dados.shape
```

```python
treino_x = x[:75]
treino_y = y[:75]

teste_x = x[75:]
teste_y = y[75:]
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)
```

```python
taxa_acerto = accuracy_score(teste_y, previsoes)
print("Taxa de acerto: %.2f%%" % (taxa_acerto*100))
```

```python
## Para nao precisar fazer na mao a divisão de treino e teste:
from sklearn.model_selection import train_test_split

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25, stratify = y) # divisao aleatório de treino e teste
# se nao quiser aleatorio, seta a SEED:
# SEED = 20
# treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, test_size = 0.25)
# stratify: manter a proporção entre treino e teste

modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

taxa_acerto = accuracy_score(teste_y, previsoes)
print("Taxa de acerto: %.2f%%" % (taxa_acerto*100))
```
